```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("traffic_accidents.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Convert crash_date to datetime format with a specified format
df['crash_date'] = pd.to_datetime(df['crash_date'], format='%Y-%m-%d')  # Adjust format as needed

# Extract year, month, and day of the week for aggregation
df['year'] = df['crash_date'].dt.year
df['month'] = df['crash_date'].dt.month_name()
df['crash_day_of_week'] = df['crash_date'].dt.day_name()

# Count the number of accidents that occurred on each day of the week
daily_accident_counts = df['crash_day_of_week'].value_counts()

# Plotting the daily accident counts
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_accident_counts.index, y=daily_accident_counts.values, hue=daily_accident_counts.index, palette="viridis", legend=False)
plt.title("Total Accidents by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.grid(axis='y')
plt.show()

# Display the updated dataframe structure
print("Total number of accidents by day of the week:")
print(daily_accident_counts)

# Display the updated dataframe structure
df[['crash_date', 'year', 'month', 'crash_day_of_week']].head()
```

```python
# Assuming you have already extracted the hour from the crash_date
df['crash_hour'] = df['crash_date'].dt.hour

# Count the number of accidents for each hour
hourly_accident_counts = df['crash_hour'].value_counts().sort_index()

# Plotting the hourly accident counts
plt.figure(figsize=(12, 6))
sns.barplot(x=hourly_accident_counts.index, y=hourly_accident_counts.values, hue=hourly_accident_counts.index, palette="viridis", legend=False)
plt.title("Total Accidents by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.xticks(hourly_accident_counts.index)  # Show each hour on the x-axis
plt.grid(axis='y')
plt.show()
```

```python
# Assuming monthly_yearly_counts is your DataFrame for months and years
# Convert the counts to integers
monthly_yearly_counts = monthly_yearly_counts.fillna(0).astype(int)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(monthly_yearly_counts, cmap="YlGnBu", annot=True, fmt='d', linewidths=.5)
plt.title("Accidents by Month and Year")
plt.ylabel("Year")
plt.xlabel("Month")
plt.show()
```

```python
# Group by year and day of the week
day_year_counts = df.groupby(['year', 'crash_day_of_week']).size().unstack()

# Plot grouped bar chart
day_year_counts.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title("Accidents by Day of the Week and Year")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.legend(title='Day of the Week')
plt.show()
```

```python
# Count total accidents per year
yearly_counts = df['year'].value_counts().sort_index()

# Plot line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o', color='g')
plt.xlabel("Year")
plt.ylabel("Total Number of Accidents")
plt.title("Total Accidents by Year")
plt.grid()
plt.show()
```
